# Puzzle Themes vs. Popularity<a name='theme'></a>

Below, we process the `Themes` data from `puzzles_df` as related to `Popularity` in a similar manner to what we did with `Themes` and `Rating` in the Puzzle Themes vs. Puzzle Rating section.

```python
theme_categories = ['advancedPawn', 'advantage', 'anastasiaMate', 'arabianMate', 'attackingF2F7',
                    'attraction', 'backRankMate', 'bishopEndgame', 'bodenMate', 'castling',
                    'capturingDefender', 'crushing', 'doubleBishopMate', 'dovetailMate', 'equality',
                    'kingsideAttack', 'clearance', 'defensiveMove', 'deflection', 'discoveredAttack',
                    'doubleCheck', 'endgame', 'enPassant', 'exposedKing', 'fork',
                    'hangingPiece', 'hookMate', 'interference', 'intermezzo', 'knightEndgame', 
                    'long', 'master', 'masterVsMaster', 'mate', 'mateIn1', 
                    'mateIn2', 'mateIn3', 'mateIn4', 'mateIn5', 'middlegame',
                    'oneMove', 'opening', 'pawnEndgame', 'pin', 'promotion',
                    'queenEndgame', 'queenRookEndgame', 'queensideAttack', 'quietMove', 'rookEndgame',
                    'sacrifice', 'short', 'skewer', 'smotheredMate', 'superGM', 
                    'trappedPiece', 'underPromotion', 'veryLong', 'xRayAttack', 'zugzwang']
```

```python
def theme_filter(theme):
  #Set the search condition in `mask` to look for the string `theme` in the `Themes` column.
  mask = puzzles_df['Themes'].str.contains(theme, case = True, na = False, regex = False)
  #Return the dataframe containing only puzzles with the specified `theme` in `Themes` column. 
  #We cast `Opening_Tags` as a `string` to avoid encoding errors when passing to `R`.
  return puzzles_df[mask].astype({'Opening_Tags': 'str'})
```

```python
theme_popularity_list = []
for theme in theme_categories:
  theme_popularity_list.append(theme_filter(theme)['Popularity'].tolist())
```

```python
theme_popularity_dictionary = dict(zip(theme_categories, theme_popularity_list))
```

```python
theme_popularity_df = pd.DataFrame.from_dict(theme_popularity_dictionary, orient = 'index')
theme_popularity_df = theme_popularity_df.transpose()
theme_popularity_df.head()
```

```python
melted_theme_popularity_df = theme_popularity_df.melt(id_vars = None, 
                                                var_name = 'Theme', 
                                                value_name = 'Popularity')

melted_theme_popularity_df.head()
```

We now save the melted data from for use with `R`.

```python
melted_theme_popularity_df.to_csv('melted_theme_popularity.csv')
```

# Opening vs. Popularity<a name='opening'></a>

Below, we process the `Opening_Tags` data from `puzzles_df` with respect to `Popularity` in a similar manner to what we did with `Opening_Tags` and `Rating` in `lichess_db_puzzle_eda_rating.ipynb`.

```python
opening_puzzles_df = puzzles_df.dropna()

opening_popularity_df = opening_puzzles_df[['Popularity','Opening_Tags']]

opening_popularity_df['Opening_Tags'] = opening_popularity_df['Opening_Tags'].map(lambda x: x.split()[0])

opening_popularity_df.head()
```

```python
%Rpush opening_popularity_df
```

```python
%%R -h 16 -w 8 -u in
ggplot(opening_popularity_df, 
       aes(y = reorder(Opening_Tags, Popularity, FUN = median, na.rm = TRUE), x = Popularity))+
  labs(y = "Opening")+
  geom_boxplot()+
  theme_minimal()
```

Median `Popularity` does not vary much with `Opening` overall. The opening with lowest median popularity is the `Queens_Pawn_Mengarini_Attack`.
